```python
import pandas as pd, sklearn
```

```python
frame = pd.read_csv("01_abalone.csv")
frame
```

```python
# CHECKING FOR NULL VALUES
frame.isnull().sum()
```

```python
# CHECKING IF ANY CATEGORICAL COLUMNS EXIST
# (THOUGH HERE WE CAN VISIBLY VERIFY THAT ONLY ONE COLUMM EXISTS)

categorical_columns = frame.select_dtypes(exclude=['number']).columns
categorical_columns
```

```python
# CHECKING THE TYPES OF THE "Sex" COLUMNS
frame["Sex"].value_counts()
```

```python
 # REPLACING WITH NUMBERS
NumberAssign = {"M":1, "I":0, "F":-1}
frame["Sex"].replace( NumberAssign, inplace = True )
```

```python
frame["Sex"].value_counts()
```

#### Seperating the features and the target

```python
x = frame.drop("Rings", axis = 1)
y = frame["Rings"]
```

```python
x
```

```python
y
```

#### Split, Train, Test

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split( x, y, test_size=0.25, random_state=50)  
```

```python
len(xtrain), len(ytrain), len(xtest), len(ytest)
```

#### Training

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

#### Results

```python
# checking the R2 score
lr.score(xtest, ytest)
```

```python
# checking the values of the errors
ytest - lr.predict( xtest )
```
